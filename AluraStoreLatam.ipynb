### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
try:
    df_tienda_1 = tienda
    df_tienda_2 = tienda2
    df_tienda_3 = tienda3
    df_tienda_4 = tienda4

    print("¡Dataframes de las 4 tiendas cargados exitosamente!")

except NameError:
    print("--- ¡ERROR! ---")
    print("No se encontraron los dataframes de las tiendas.")
    print("Por favor, asegúrate de haber ejecutado la primera celda para cargar los dataframes desde las URLs.")
    raise

COL_PRECIO = 'Precio'

try:
    fact_1 = df_tienda_1[COL_PRECIO].sum()
    fact_2 = df_tienda_2[COL_PRECIO].sum()
    fact_3 = df_tienda_3[COL_PRECIO].sum()
    fact_4 = df_tienda_4[COL_PRECIO].sum()

    print("\n--- Facturación Total Calculada ---")
    print(f"Tienda 1: ${fact_1:,.0f}")
    print(f"Tienda 2: ${fact_2:,.0f}")
    print(f"Tienda 3: ${fact_3:,.0f}")
    print(f"Tienda 4: ${fact_4:,.0f}")

except KeyError:
    print(f"--- ¡ERROR! ---")
    print(f"No se encontró la columna '{COL_PRECIO}' en uno de los dataframes.")
    print("Por favor, verifica que el nombre de la columna es correcto.")
    raise
except Exception as e:
    print(f"Ocurrió un error inesperado al calcular: {e}")
    raise

datos_facturacion = {
    'Tienda 1': fact_1,
    'Tienda 2': fact_2,
    'Tienda 3': fact_3,
    'Tienda 4': fact_4
}

series_facturacion = pd.Series(datos_facturacion).sort_values()

plt.figure(figsize=(10, 6))

grafico = sns.barplot(
    x=series_facturacion.index,
    y=series_facturacion.values,
    palette='Blues_d'
)


if grafico.patches: # Check if patches exist before trying to access index 0
    grafico.patches[0].set_facecolor('red')


plt.title('Análisis de Facturación Total por Tienda', fontsize=16, fontweight='bold')
plt.ylabel('Facturación Total (COP $)', fontsize=12)
plt.xlabel('Tienda', fontsize=12)

for p in grafico.patches:
    grafico.annotate(
        f'${p.get_height():,.0f}', # Texto a mostrar (formato de moneda)
        (p.get_x() + p.get_width() / 2., p.get_height()), # Posición
        ha='center',
        va='center',
        xytext=(0, 9), # Desplazamiento vertical
        textcoords='offset points',
        fontweight='bold'
    )

plt.savefig("analisis_facturacion.png")
print("\nGráfico 'analisis_facturacion.png' guardado exitosamente.")

plt.show()
```

# 2. Valoracion Media por Tienda

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

try:
    df_tienda_1 = tienda
    df_tienda_2 = tienda2
    df_tienda_3 = tienda3
    df_tienda_4 = tienda4

    print("¡Dataframes de las 4 tiendas cargados exitosamente!")

except NameError:
    print("--- ¡ERROR! ---")
    print("No se encontraron los dataframes de las tiendas.")
    print("Por favor, asegúrate de haber ejecutado la primera celda para cargar los dataframes desde las URLs.")

    raise



COL_VALORACION = 'Calificación'

try:

    val_1 = df_tienda_1[COL_VALORACION].mean()
    val_2 = df_tienda_2[COL_VALORACION].mean()
    val_3 = df_tienda_3[COL_VALORACION].mean()
    val_4 = df_tienda_4[COL_VALORACION].mean()

    print("\n--- Valoración Promedio Calculada ---")
    print(f"Tienda 1: {val_1:.2f} estrellas")
    print(f"Tienda 2: {val_2:.2f} estrellas")
    print(f"Tienda 3: {val_3:.2f} estrellas")
    print(f"Tienda 4: {val_4:.2f} estrellas")

except KeyError:
    print(f"--- ¡ERROR! ---")
    print(f"No se encontró la columna '{COL_VALORACION}' en uno de los dataframes.")
    print("Por favor, verifica que el nombre de la columna es correcto.")
    raise
except Exception as e:
    print(f"Ocurrió un error inesperado al calcular: {e}")
    raise


datos_valoracion = {
    'Tienda 1': val_1,
    'Tienda 2': val_2,
    'Tienda 3': val_3,
    'Tienda 4': val_4
}


series_valoracion = pd.Series(datos_valoracion).sort_values(ascending=True)



plt.figure(figsize=(10, 6))


grafico = sns.barplot(
    x=series_valoracion.index,
    y=series_valoracion.values,
    palette='OrRd'
)


if grafico.patches:
    grafico.patches[0].set_facecolor('darkred')

plt.title('Análisis de Valoración Promedio por Tienda', fontsize=16, fontweight='bold')
plt.ylabel('Valoración Promedio (Estrellas)', fontsize=12)
plt.xlabel('Tienda', fontsize=12)


plt.ylim(0, 5.5)

promedio_general = series_valoracion.mean()
plt.axhline(y=promedio_general, color='gray', linestyle='--', label=f'Promedio Gral. ({promedio_general:.2f} est.)')
plt.legend()

for p in grafico.patches:
    grafico.annotate(
        f'{p.get_height():.2f}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='center',
        xytext=(0, 9),
        textcoords='offset points',
        fontweight='bold'
    )

plt.savefig("analisis_valoracion.png")
print("\nGráfico 'analisis_valoracion.png' guardado exitosamente.")

plt.show()
```

# 4. Productos más y menos vendidos


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

try:
    df_tienda_1 = tienda
    df_tienda_2 = tienda2
    df_tienda_3 = tienda3
    df_tienda_4 = tienda4

    print("¡Dataframes de las 4 tiendas cargados exitosamente!")

except NameError:
    print("--- ¡ERROR! ---")
    print("No se encontraron los dataframes de las tiendas.")
    print("Por favor, asegúrate de haber ejecutado la primera celda para cargar los dataframes desde las URLs.")
    raise



COL_PRODUCTO = 'Producto'


print("\n--- Resumen de Productos Más/Menos Vendidos por Tienda ---")

try:

    def analizar_productos_tienda(df, nombre_tienda):
        ventas_producto = df[COL_PRODUCTO].value_counts()

        mas_vendido = ventas_producto.index[0]
        ventas_mas = ventas_producto.iloc[0]

        menos_vendido = ventas_producto.index[-1]
        ventas_menos = ventas_producto.iloc[-1]

        print(f"--- {nombre_tienda} ---")
        print(f"  Más vendido: {mas_vendido} ({ventas_mas} ventas)")
        print(f"  Menos vendido: {menos_vendido} ({ventas_menos} ventas)")

    analizar_productos_tienda(df_tienda_1, "Tienda 1")
    analizar_productos_tienda(df_tienda_2, "Tienda 2")
    analizar_productos_tienda(df_tienda_3, "Tienda 3")
    analizar_productos_tienda(df_tienda_4, "Tienda 4")

except KeyError:
    print(f"--- ¡ERROR! ---")
    print(f"No se encontró la columna '{COL_PRODUCTO}'.")
    print("Por favor, verifica que el nombre de la columna es correcto.")
    raise
except Exception as e:
    print(f"Ocurrió un error inesperado al calcular: {e}")
    raise


df_global = pd.concat([df_tienda_1, df_tienda_2, df_tienda_3, df_tienda_4], ignore_index=True)


ventas_globales = df_global[COL_PRODUCTO].value_counts()


top_5_productos = ventas_globales.head(5)

plt.figure(figsize=(12, 7))

grafico_top = sns.barplot(
    x=top_5_productos.values,
    y=top_5_productos.index,
    palette='viridis'
)

plt.title('Top 5 Productos MÁS Vendidos (Global - Todas las Tiendas)', fontsize=16, fontweight='bold')
plt.xlabel('Cantidad Total de Ventas', fontsize=12)
plt.ylabel('Producto', fontsize=12)


for p in grafico_top.patches:
    grafico_top.annotate(
        f'{int(p.get_width())}',
        (p.get_width(), p.get_y() + p.get_height() / 2.),
        ha='left', va='center',
        xytext=(5, 0),
        textcoords='offset points',
        fontweight='bold'
    )

plt.tight_layout()
plt.savefig("analisis_top_5_productos.png")
print("\nGráfico 'analisis_top_5_productos.png' guardado exitosamente.")
plt.show()

bottom_5_productos = ventas_globales.tail(5).sort_values(ascending=True)

plt.figure(figsize=(12, 7))
grafico_bottom = sns.barplot(
    x=bottom_5_productos.values,
    y=bottom_5_productos.index,
    palette='rocket_r'
)
plt.title('Top 5 Productos MENOS Vendidos (Global - Todas las Tiendas)', fontsize=16, fontweight='bold')
plt.xlabel('Cantidad Total de Ventas', fontsize=12)
plt.ylabel('Producto', fontsize=12)


for p in grafico_bottom.patches:
    grafico_bottom.annotate(
        f'{int(p.get_width())}',
        (p.get_width(), p.get_y() + p.get_height() / 2.),
        ha='left', va='center',
        xytext=(5, 0),
        textcoords='offset points',
        fontweight='bold'
    )

plt.tight_layout()
plt.savefig("analisis_bottom_5_productos.png")
print("Gráfico 'analisis_bottom_5_productos.png' guardado exitosamente.")
plt.show()
```

Gráfico 3: Análisis de Costo Promedio de Envío por Tienda

Por el contrario, la Tienda 4 es la que menos contribuye financieramente a la cadena, convirtiéndola en la candidata lógica para ser liquidada y permitir así la reinversión de ese capital.
